```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('Unemployment in India.csv')
```

```python
print(data.info())
print(data.head())
```

```python
print("Missing Values:\n", data.isnull().sum())
print("Summary Statistics:\n", data.describe())
```

```python
print(data.columns)
data.rename(columns=lambda x: x.strip(), inplace=True)  # Removes leading/trailing spaces
data.rename(columns={' date': 'Date'}, inplace=True)
print(data.head())
if 'Date' in data.columns:
    data['Date'] = pd.to_datetime(data['Date'])
else:
    print("Error: 'Date' column not found in the dataset.")
```

```python
data['Date'] = pd.to_datetime(data['Date'])
```

```python
plt.figure(figsize=(12, 6))
sns.lineplot(x='Date', y='Estimated Unemployment Rate (%)', hue='Region', data=data)
plt.title('Unemployment Rate Over Time by Region')
plt.xlabel('Date')
plt.ylabel('Unemployment Rate (%)')
plt.legend(title='Region')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(10, 5))
sns.histplot(data['Estimated Employed'], bins=30, kde=True)
plt.title('Distribution of Estimated Employed Individuals')
plt.xlabel('Estimated Employed')
plt.ylabel('Frequency')
plt.show()
```

```python
cleaned_data_path = 'unemployment_analysis.csv'
data.to_csv(cleaned_data_path, index=False)
print(f"Cleaned data saved to {cleaned_data_path}")
```
